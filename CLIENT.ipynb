```python
# Echo client program
import sys
def warn(*args, **kwargs): pass
import warnings
from random import randint
import random
warnings.warn = warn

def check(s):
    if s == '404':
        sys.exit('Щось пішло не так('+'Давай перезапускай і надалі вводь коректні дані! Думай коли відповідаєш!')
import socket

HOST = 'localhost'    # The remote host
PORT = 50008              # The same port as used by the server
with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.connect((HOST, PORT))
    iter1 = s.recv(1024)
    print(iter1.decode("utf-8"))
    while True:
        try:
            cat_number = int(input("Я обираю категорію: "))
            break
        except ValueError:
            print("Упс... хтось тут ввів не те, що треба")
    s.sendall(str(cat_number).encode("utf-8"))
    ####################################
    iter2 = s.recv(1024)
    iter2 = iter2.decode("utf-8")
    check(iter2)
    print(iter2)
    ####################################
    key=1
    ans = 'ff'
    while (key==1):
        itern = s.recv(1024)
        itern = itern.decode("utf-8")
        if itern == 'end': 
            break;
        else:
            check(itern)
            print(itern)
            while True:
                try:
                    ans = int(input("Я обираю слово під номером: "))
                    break
                except ValueError:
                    print("Упс... хтось тут ввів не те, що треба")
            
            s.sendall(str(ans).encode("utf-8"))
            iternn = s.recv(1024)
            iternn = iternn.decode("utf-8")
            print(iternn)
    iterf = s.recv(1024)
    iterf = iterf.decode("utf-8")
    print(iterf)
    s.close()        
```
